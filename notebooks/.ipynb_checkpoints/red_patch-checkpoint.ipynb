```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
bgr = cv2.imread('sat1.png')
img = cv2.imread('sat1.png',0)
plt.imshow(img,cmap='gray')
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(img)
plt.imshow(cl1,cmap='gray')
```

```python
edged = cv2.Canny(bgr,50,100)
```

```python
cnts, hier = cv2.findContours(edged, cv2.RETR_EXTERNAL,
                        cv2.CHAIN_APPROX_SIMPLE)
```

```python
cnt_img = cv2.drawContours(bgr,cnts,-1,(0,255,0),1)
plt.imshow(cnt_img)
```

```python
minPerimeterRate = 0.03
maxPerimeterRate = 4.0
minPerimeter = minPerimeterRate*max(img.shape)
maxPerimeter = maxPerimeterRate*max(img.shape)
candidate_cnts = []
for cnt in cnts:
    if cv2.arcLength(cnt,True)>minPerimeter and cv2.arcLength(cnt,True)<maxPerimeter:
        candidate_cnts.append(cnt)
```

```python
cnt_img = cv2.drawContours(bgr,candidate_cnts,-1,(0,255,0),1)
plt.imshow(cnt_img)
```

```python
convex_cnts = []
for cnt in cnts:
    convex_cnts.append(cv2.convexHull(cnt))
```

```python
bgr = cv2.imread('sat1.png')
cnt_img = cv2.drawContours(bgr,convex_cnts,-1,(0,255,0),1)
plt.imshow(cnt_img)
```

```python
area_cnts = sorted(convex_cnts, key=cv2.contourArea,reverse=True)
```

```python
bgr = cv2.imread('sat1.png')
cnt_img = cv2.drawContours(bgr,area_cnts,0,(0,255,0),1)
plt.imshow(cnt_img)
```

```python
def KMeansSeg(img,k):
    if len(img.shape)>2:
        vectorized = img.reshape((-1,3))
    else:
        vectorized = img.reshape((-1,1))
    vectorized = np.float32(vectorized)
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    K = k
    attempts=10
    ret,label,center=cv2.kmeans(vectorized,K,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
    center = np.uint8(center)
    res = center[label.flatten()]
    seg_img = res.reshape((img.shape))
    return seg_img
```

```python
bgr = cv2.imread('sat1.png')
seg_img = KMeansSeg(bgr,2)
plt.imshow(seg_img)
```

```python
seg_img = cv2.Canny(seg_img,50,100)
cnts, hier = cv2.findContours(seg_img, cv2.RETR_EXTERNAL,
                        cv2.CHAIN_APPROX_SIMPLE)
area_cnts = sorted(cnts, key=cv2.contourArea,reverse=True)
cnt_img = cv2.drawContours(bgr,area_cnts,0,(0,255,0),1)
plt.imshow(cnt_img)
```
