```python
inputImage = cv2.imread("design7.png")
plt.imshow(inputImage)
```

```python
dst = inputImage
#Load the dictionary that was used to generate the markers.
dictionary = cv2.aruco.Dictionary_get(cv2.aruco.DICT_4X4_50)

# Initialize the detector parameters using default values
parameters =  cv2.aruco.DetectorParameters_create()

# Detect the markers in the image
markerCorners, markerIds, rejectedCandidates = cv2.aruco.detectMarkers(dst, dictionary, parameters=parameters)
# markerCorners = np.array(markerCorners)
# markerCorners = markerCorners.astype(np.int)
```

```python
# if len(markerIds) == 4:
markerIds = markerIds.flatten()
rectangle = np.full((1,4,2),0)
for i, markerId in enumerate(markerIds):
    corners = markerCorners[i].reshape(-1,2)
    corners = corners.astype(np.int)
    corner = corners[-1+i]
    rectangle[0][markerId-1] = corner
```

```python
img = cv2.imread("design7.png")
contour_img = cv2.drawContours(img, rectangle,0, (0, 255, 0), 3)
plt.imshow(contour_img)
```
