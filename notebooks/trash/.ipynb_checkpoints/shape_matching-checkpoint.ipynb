```python
import glob
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
#read image
filename = glob.glob('./*_edited.jpg')[-1]
bgr = cv2.imread(filename)
rgb = cv2.cvtColor(bgr,cv2.COLOR_BGR2RGB)
#convert to rgb channel for display
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
# display the rgb image
plt.figure(figsize=(20,12))
# plt.axis('off')
plt.imshow(gray,cmap='gray')
```

```python
# blurred = cv2.blur(gray, (3,3))
# apply canny algo on grayscale pic to detect edges
edged = cv2.Canny(gray, 50, 100)
edged = cv2.Canny(edged, 50, 100)
plt.figure(figsize=(20,12))
plt.axis('off')
plt.imshow(edged,cmap='gray')
```

```python
kernel = np.ones((5,5), np.uint8) 
img_dilation = cv2.dilate(edged, kernel, iterations=1) 
img_erosion = cv2.erode(img_dilation, kernel, iterations=1) 
plt.figure(figsize=(20,12))
plt.axis('off')
plt.imshow(img_dilation,cmap='gray')
img = img_erosion
```

```python
blurred = cv2.medianBlur(edged.copy(), 5)
src = blurred
rows = src.shape[0]
circles = cv2.HoughCircles(src, cv2.HOUGH_GRADIENT, 1, minDist = rows/8,
                               param1=100, param2=50,
                               minRadius=50, maxRadius=500)
# h,w = src.shape
drawing = rgb.copy()
cir_img = drawing
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        center = (i[0], i[1])
        # circle center
        cir_img = cv2.circle(drawing, center, 1, (0, 255, 0), 3)
        # circle outline
        radius = i[2]
        cir_img = cv2.circle(cir_img, center, radius, (0, 0, 255), 3)
plt.figure(figsize=(20,12))
plt.imshow(cir_img)
```

```python
import math
src = gray
lines = cv2.HoughLines(edged, 1, np.pi / 180, 150, None, 0, 0)
    
if lines is not None:
    for i in range(0, len(lines)):
        rho = lines[i][0][0]
        theta = lines[i][0][1]
        a = math.cos(theta)
        b = math.sin(theta)
        x0 = a * rho
        y0 = b * rho
        pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
        pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))
        line_img = cv2.line(src, pt1, pt2, (0,0,255), 3, cv2.LINE_AA)
plt.figure(figsize=(20,12))
plt.axis('off')
plt.imshow(line_img)
```
